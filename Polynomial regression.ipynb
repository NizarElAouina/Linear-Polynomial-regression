```python
def afficher_pol(LPol):
    L=LPol[:]; L.reverse()
    L=[round(e,2) for e in L]
    for i in range(len(L)):
        print('('+str(L[i])+'*X^'+str(i)+')',end='+')
```

```python
def Evaluer_pol(LPol,x):
    s=0
    L=LPol[:]
    L.reverse()
    for i in range(len(LPol)):
        s+=L[i]*(x**i)
    return(s)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
data=pandas.read_csv("C:/Users/nizar/Desktop/S2/Machine Learning/TP 2/dataset_polynomial3.csv").values
data
```

```python
x=data[1:1001,0]
y=data[1:1001,1]
degre_polynome=2
p= np.poly1d(np.polyfit(x, y,degre_polynome)) #retourne un vecteur de coefficients de polynôme ordonnés par ordre décroissant de degré
p
```

```python
LPol=list(p)
afficher_pol(LPol)
les_x= np.linspace(min(x), max(x), 100)
plt.scatter(x, y)
plt.plot(les_x, p(les_x), c='r')
plt.show()
#Evaluer_pol(LPol,x)
```
