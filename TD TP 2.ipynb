```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
df = pd.read_csv("C:/Users/nizar/Desktop/S2/Machine Learning/TP 2/univariate_linear_regression_dataset.csv")
X = df.iloc[:,0]
Y = df.iloc[:,1]
X
```

```python
Y
```

```python
axes = plt.axes()
```

```python
axes.grid()
plt.scatter(X,Y)
```

```python
plt.show()
SL=stats.linregress(X, Y)
slope=SL.slope; intercept=SL.intercept; coef_correlation=SL.rvalue
coef_correlation
```

```python
def predict(x):
    return slope * x + intercept
```

```python
axes = plt.axes()
axes.grid()
```

```python
plt.scatter(X,Y)
fitLine = predict(X)
plt.plot(X, fitLine, c='r')
plt.show()
print(slope)
print(intercept)
print(coef_correlation)
```
